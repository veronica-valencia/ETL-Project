```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
sp500_file = "sp500.csv"
sp500_df = pd.read_csv(sp500_file)
sp500_df.head()
cpi_file = "cpi_data.csv"
cpi_df = pd.read_csv(cpi_file)
cpi_df.head()
housing_file = "housing.csv"
housing_df = pd.read_csv(housing_file)
housing_df.head()
consumer_spending_file = "consumerspending.csv"
consumer_spending_df = pd.read_csv(consumer_spending_file)
consumer_spending_df.head()
housing_df.head()
unemployment_file = "unemployment.csv"
unemployment_df = pd.read_csv(unemployment_file)
unemployment_df.head()
spending_sentiment_file = "spending_sentiment.csv"
spending_sentiment_df = pd.read_csv(spending_sentiment_file)
spending_sentiment_df.head()
```

### Transform DataFrames

```python
sp500_cols = ['Date','Close','Adj Close','Volume']
sp500_transformed= sp500_df[sp500_cols].copy()

# Rename the column headers
sp500_transformed = sp500_transformed.rename(columns={"Date": "date",
                                                            "Close": "close",
                                                          "Adj Close": "adj_close","Volume": "volume"})

# Clean the data by dropping duplicates and setting the index
sp500_transformed.drop_duplicates("date", inplace=True)
#sp500_transformed.set_index("date", inplace=True)

sp500_transformed.head()


housing_cols = ['DATE', 'CSUSHPISA']
housing_transformed= housing_df[housing_cols].copy()

# Rename the column headers
housing_transformed = housing_transformed.rename(columns={"DATE": "date","CSUSHPISA": "csushpisa"})

# Clean the data by dropping duplicates and setting the index
housing_transformed.drop_duplicates("date", inplace=True)
#housing_transformed.set_index("date", inplace=True)

housing_transformed.head()


consumer_spending_cols = ['DATE', 'PCE']
consumer_spending_transformed= consumer_spending_df[consumer_spending_cols].copy()

# Rename the column headers
consumer_spending_transformed = consumer_spending_transformed.rename(columns={"DATE": "date","PCE": "pce"})

# Clean the data by dropping duplicates and setting the index
consumer_spending_transformed.drop_duplicates("date", inplace=True)
#consumer_spending_transformed.set_index("date", inplace=True)

consumer_spending_transformed.head()

cpi_cols = ['DATE', 'CPIAUCSL']
cpi_transformed= cpi_df[cpi_cols].copy()

# Rename the column headers
cpi_transformed = cpi_transformed.rename(columns={"DATE": "date","CPIAUCSL": "cpi_aucsl"})

# Clean the data by dropping duplicates and setting the index
cpi_transformed.drop_duplicates("date", inplace=True)
#cpi_transformed.set_index("date", inplace=True)

cpi_transformed.head()


unemployment_cols = ['DATE', 'UNRATE']
unemployment_transformed= unemployment_df[unemployment_cols].copy()

# Rename the column headers
unemployment_transformed = unemployment_transformed.rename(columns={"DATE": "date","UNRATE": "unrate"})

# Clean the data by dropping duplicates and setting the index
unemployment_transformed.drop_duplicates("date", inplace=True)
#unemployment_transformed.set_index("date", inplace=True)

unemployment_transformed.head()

spending_sentiment_cols = ['DATE', 'UMCSENT']
spending_sentiment_transformed= spending_sentiment_df[spending_sentiment_cols].copy()

# Rename the column headers
spending_sentiment_transformed = spending_sentiment_transformed.rename(columns={"DATE": "date","UMCSENT": "umcsent"})

# Clean the data by dropping duplicates and setting the index
spending_sentiment_transformed.drop_duplicates("date", inplace=True)
#spending_sentiment_transformed.set_index("date", inplace=True)

spending_sentiment_transformed.head()
```

```python
sp500_transformed = sp500_transformed[['date', 'close']]

sp500_transformed2 = sp500_transformed.rename(columns={'close': 'Average_Close'}) 

sp500_transformed2.head()
```

```python
combine_df = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(sp500_transformed2,housing_transformed, on = 'date', how = 'left' ), cpi_transformed, on = 'date', how = 'left'),
                      consumer_spending_transformed, on = 'date', how = 'left'),unemployment_transformed, on = 'date', how = 'left'),spending_sentiment_transformed,on = 'date',how = 'left')

combine_df.head()
```

```python
Final_df = combine_df.rename(columns={'Average_Close': 'Average_SP500', 'csushpisa':'HPI', 'cpi_aucsl':'CPI', 'pce': 'CS', 'unrate':'UNRate', 'umcsent': 'CSent'})

Final_df.head()
```

### Load the Data

```python
#postgres 
import psycopg2

# SQL Alchemy
from sqlalchemy import create_engine

from sqlalchemy import MetaData
```

```python
# Connect to local Database

rds_connection_string = "postgres:postgres@localhost:5432/economy_db"

engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Use pandas to load csv converted DataFrame into database

sp500_transformed2.to_sql(name='sp500', con=engine, if_exists='replace', index=False)

housing_transformed.to_sql(name='housing', con=engine, if_exists='replace', index=False)

cpi_transformed.to_sql(name='cpi', con=engine, if_exists='replace', index=False)

consumer_spending_transformed.to_sql(name='spending', con=engine, if_exists='replace', index=False)

unemployment_transformed.to_sql(name='unemployment', con=engine, if_exists='replace', index=False)

spending_sentiment_transformed.to_sql(name='sentiment', con=engine, if_exists='replace', index=False)

# Upload the Merged data frame 

Final_df.to_sql(name='combined', con=engine, if_exists='replace', index=False)
```
